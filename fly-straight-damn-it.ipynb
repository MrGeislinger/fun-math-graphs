```python
import numpy as np
import matplotlib.pyplot as plt
```

- if gcd = 1, n + a(n-1) + 1  (not co prime)
- if gcd > 1, a(n-1) / gcd()   

```python
# seqeunce = [1,1,4,8]
# Seed needs to have first two defined
sequence = [1,1]

n=2
```

```python
# is next good?
for i in range(900):
    gcd = np.gcd(sequence[n-1],n)
    if gcd == 1:
        next_number = 1 + n + sequence[n-1]
    else:
        next_number = int(sequence[n-1] / gcd)
    n += 1
    sequence.append(next_number)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(range(len(sequence)),sequence, alpha=0.4, s=15);
```
